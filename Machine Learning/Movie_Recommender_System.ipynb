```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
# Merging the dataset
movies = movies.merge(credits,on='title')
```

```python
# Considering only concerned features
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
print("Check for NULL values :\n",movies.isnull().sum())
print("\nDuplicated Data : ",movies.duplicated().sum())
```

```python
import ast
```

```python
def cnvrt_strlit(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
def cnvrt_crew(obj):
  L = []
  count = 0
  for i in ast.literal_eval(obj):
    L.append(i['name'])
    if(count<2):
      count = count + 1
    else:
      break
  return L
```

```python
movies['genres'] = movies['genres'].apply(cnvrt_strlit)
```

```python
movies['keywords'] = movies['keywords'].apply(cnvrt_strlit)
```

```python
movies['cast'] = movies['cast'].apply(cnvrt_crew)
```

```python
movies['cast']
```

```python
movies.iloc[0].cast
```

```python
def fetch_director(obj):
  Director_name = []
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      Director_name.append(i['name'])
      return Director_name
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])

movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tag'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tag']]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tag']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  L = []
  for i in text.split():
    L.append(ps.stem(i))
  return " ".join(L)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```
