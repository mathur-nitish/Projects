```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/content/50_Startups.csv')
```

```python
dataset.head()
```

```python
print("State values : ",dataset['State'].nunique())
print("\nMissing Values")
print(dataset.isnull().sum())
print("\nNo. of  Rows:",dataset.shape[0])
print(dataset.count())
print("\nNo. of Columns:",dataset.shape[1])
print(dataset.columns)
```

```python
# One Hot Encoding
dataset_new = pd.get_dummies(dataset,columns=['State'],drop_first=True)
```

```python
dataset_new.head()
```

```python
# splitting variables
x = dataset_new.drop('Profit',axis=1)
```

```python
y = dataset_new['Profit']
```

```python
print("Independent Variable (x) ")
print(x)
print("\nDependent Variable (y) ")
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
```

```python
y_pred.reshape((y_pred.shape[0],1))
```

```python
y_testnp = y_test.to_numpy()
```

```python
# Comparison of Predicted and Real Values
np.concatenate((y_pred.reshape((y_pred.shape[0],1)), y_testnp.reshape((len(y_testnp),1))),1)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
```

```python
r_squared = r2_score(y_test, y_pred)

# Compute adjusted R-squared (requires number of features)
n = len(y_train)
# Number of features
p = x_train.shape[1]
adjusted_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)
print("R-squared:", r_squared)
print("Adjusted R-squared:", adjusted_r_squared)
```
